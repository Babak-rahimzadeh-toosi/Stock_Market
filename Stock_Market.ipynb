```python
import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as ctrl
```

```python
MACD = ctrl.Antecedent(np.arange(0,11,0.1), "MACD")
```

```python
RSI = ctrl.Antecedent(np.arange(0,110,0.1), "RSI")
```

```python
SO = ctrl.Antecedent(np.arange(0,101,0.1), "SO")
```

```python
OBV = ctrl.Antecedent(np.arange(0,11,0.1), "OBV")
```

```python
RECOMMENDATION = ctrl.Consequent(np.arange(0, 35, 5), "RECOMMENDATION", defuzzify_method = "centroid")
```

```python
RECOMMENDATION["SELL"] = fuzzy.trapmf(RECOMMENDATION.universe, [0, 0, 5, 10])
```

```python
RECOMMENDATION["HOLD"] = fuzzy.trimf(RECOMMENDATION.universe, [5, 10, 15])
```

```python
RECOMMENDATION["BUY"] = fuzzy.trapmf(RECOMMENDATION.universe, [10, 15, 31, 31])
```

```python
Market_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10,
                                  rule11, rule12, rule13, rule14, rule15, rule16, rule17, rule18, rule19, rule20,
                                  rule21, rule22, rule23, rule24, rule25, rule26, rule27, rule28, rule29, rule30,
                                  rule31, rule32, rule33, rule34, rule35, rule36])
Market = ctrl.ControlSystemSimulation(Market_ctrl)
```

```python
input_data = [
    {'MACD': 6.5, 'RSI': 50, 'SO': 70, 'OBV': 2 },
    {'MACD': 3.6, 'RSI': 32, 'SO': 22, 'OBV': 7},
    {'MACD': 8.0, 'RSI': 90, 'SO': 45, 'OBV': 9 },
    {'MACD': 1.5, 'RSI': 20, 'SO': 36, 'OBV': 3},
    {'MACD': 7.0, 'RSI': 89, 'SO': 63, 'OBV': 8},
]

for data in input_data:
    Market.input['MACD'] = data['MACD']
    Market.input['RSI'] = data['RSI']
    Market.input['SO'] = data['SO']
    Market.input['OBV'] = data['OBV']

    Market.compute()

    print(f"Input (MACD={data['MACD']}, RSI={data['RSI']}, SO={data['SO']}, OBV={data['OBV']}): Output RECOMMENDATION={Market.output['RECOMMENDATION']:.2f}")
```
